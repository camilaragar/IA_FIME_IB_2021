1.1 El usuario dará un número al azar y el código calculará la suma de todos los números desde el 1 hasta el número dado por el usuario.
Ejemplo: el usuario dió el número 4, el código deberá de dar como resultado 10.

```python
# se pide el número al usuario
numero = int (input("Proporcione un número: ")) 

#declaramos el contador de suma y de índice
suma=0
i=1
#ahora se procederá a utilizar un for loop para iterar por cada número
for i in range(0,numero+1):
#dento del for loop, tenemos el valor de la suma que actuará como nuestro contador 
#y cada vez que "i" cambie su valor, se le agregará al valor total de la suma, y al final se incrementará el valor de "i"
  suma= suma + i
  i = i + 1
print("la suma de todos los numeros hasta el", numero, "es: ", suma)
```

1.2 Dados el inicio y final de un rango de números, guardar ese rango de números en una lista. Después, imprimir los números que son pares en la lista por medio de uno de los ciclos que vimos en clase. 
Inicio = 6, final = 31

```python

#primero crearemos nuestra lista con la función de python range, dándole los limites y el step de valor 1
#el asterisco en la función de range nos hará un desempaquetado de los números
lista=[*range(6,31+1,1)]
print(lista)
#mensaje para distinguir la lista principal de la lista de números pares
print("lista con solo numeros pares: ")

#ahora utilizaremos un for loop y condición del operador módulo para imprimir solo los números pares
for numero in lista:
  if numero % 2 == 0:
    print(numero, end = " ")

```
